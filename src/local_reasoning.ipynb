```python
import os
from pathlib import Path
import importlib

import new_API_reason

# importlib.reload(new_API_reason)
```

```python
OPENAI_API_KEY=""  # Set your OpenAI API key here
```

```python
EXPERIMENT_NAME="exp0"
DATA_DIR="experiments"
EXPERIMENT_OUT=f"{DATA_DIR}/{EXPERIMENT_NAME}/results"


flip_challenges_dir=f"{DATA_DIR}/full_flips_set/full_val_split/tasks"
images_dir=f"{DATA_DIR}/full_flips_set/full_val_split/tasks/images"
reasoning_models="ChatGPT-o3-mini-low"
caption_models="BLIP2_flan_t5_xxl"
caption_file_paths=f"{DATA_DIR}/{EXPERIMENT_NAME}/captions/full_val_split_images__BLIP2_6_7B_COCO,BLIP2_flan_t5_xxl.json"


params_file_path=f"{DATA_DIR}/params_reason_0.json"

output_file_name=f"{reasoning_models}__{caption_models}.json"
output_file=f"{EXPERIMENT_OUT}/{output_file_name}"

TEST_MODE=0
VERBOSE=10

TASK_FORMAT="single_sentence"
context_dir = os.path.join(DATA_DIR, "context")


# Parse and validate arguments
flip_challenges_dir = Path(flip_challenges_dir) if flip_challenges_dir else None
if flip_challenges_dir is None:
    raise Exception("flip_challenges_dir is not provided")

reasoning_models = reasoning_models.split(',') if ',' in reasoning_models else [reasoning_models]
reasoning_models = [model.strip() for model in reasoning_models]

caption_models = caption_models.split(',') if ',' in caption_models else [caption_models]
caption_models = [model.strip() for model in caption_models]

caption_file_paths = caption_file_paths.split('::') if '::' in caption_file_paths else [caption_file_paths]
caption_file_paths = [path.strip() for path in caption_file_paths]

images_dir = Path(images_dir) if images_dir else None
if images_dir is None:
    raise Exception("images_dir is not provided")

params_file_path = Path(params_file_path) if params_file_path else None
if params_file_path is None:
    raise Exception("params_file_path is not provided")

output_file = Path(output_file) if output_file else None
if output_file is None:
    raise Exception("output_file is not provided")


# Configure settings
config = new_API_reason.Config()
config.openai_api_key = OPENAI_API_KEY
config.test_mode = int(TEST_MODE) if TEST_MODE is not None else 0
config.verbose = int(VERBOSE) if VERBOSE is not None else 0
config.task_format = TASK_FORMAT if TASK_FORMAT else "single_sentence"
config.context_dir = context_dir


# Create the challenger and run
challenger = new_API_reason.FlipChallenger(config)
challenger.run(
    flip_challenges_dir=flip_challenges_dir,
    reasoning_models=reasoning_models,
    caption_models=caption_models,
    caption_file_paths=caption_file_paths,
    images_dir=images_dir,
    params_file_path=params_file_path,
    output_file=output_file
)
```
